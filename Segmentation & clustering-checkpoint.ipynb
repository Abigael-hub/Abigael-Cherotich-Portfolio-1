```python
palette = {'Male': 'blue', 'Female': 'orange'}
sns.kdeplot(data=df, x='Annual Income (k$)', shade=True, hue='Gender')
plt.show()
```

```python
palette = {'Male': 'blue', 'Female': 'orange'}

columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.kdeplot(data=df, x=i, shade=True, hue='Gender', palette=palette)
    plt.title(f'KDE Plot for {i}')
    plt.xlabel(i)
    plt.ylabel('Density')
    
    # Create a legend manually
    handles = [plt.Rectangle((0,0),1,1, color=palette[label]) for label in palette]
    plt.legend(handles, palette.keys(), title='Gender')
    
    plt.show()
```

```python
palette = {'Male': 'blue', 'Female': 'orange'}
columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.boxplot(data=df, x='Gender', y=df[i],palette=palette)
```
